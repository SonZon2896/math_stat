```python
import numpy as np

x = [-1.11, -6.10, 2.42]
y = [-2.29, -2.91]
```

```python
btst_iter = 10000
betta = 0.95
h = 0
h_wave = np.mean(x) - np.mean(y)

delta = h_wave - h

btst_delta = []
for _ in range(btst_iter):
    btst_x = np.random.choice(x, size=len(x), replace=True)
    btst_y = np.random.choice(y, size=len(y), replace=True)
    btst_delta.append(np.mean(btst_x) - np.mean(btst_y) - h_wave)

variat = sorted(btst_delta)

k = int(btst_iter * (1 - beta))

upper_bound = h_wave - variat[k]

print("Доверительный интервал: h < ", upper_bound)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def F_theta(theta):
  return 1 - norm.cdf(upper_bound, loc=theta, scale=np.sqrt(3))

theta_values = np.linspace(-10, 20, 200)

F_values = [F_theta(theta) for theta in theta_values]

plt.figure(figsize=(8, 6))
plt.plot(theta_values, F_values)
plt.xlabel("θ")
plt.ylabel("W(θ)")
plt.title("График W(θ)") 
plt.grid(True)
plt.show()
```
